#VISUALIZATION_DEFINITION
Visualization is the process by which we can show our data in graphs,charts and interactive way so that user can understand data easily.
There are many types of graph and visualization in the python.
Name of some basic charts/graphs are:-
1.line graphs/charts
2.Scatter
3.Bar graph
4.Histogram
5.Pie chart
6.Box plot
There are many libraries available in python to perform visualization.the most important library is matplotlib.
Another one is Seaborn.Seaborn is the advance library of python for visualization.
We can install matplotlib library using a pip command.

#1.LINEGRAPH
Plot is used for creating a line graph.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_coordinates=np.array([0,6])
y_coordinates=np.array([0,300])

plt.plot(x_coordinates,y_coordinates)
plt.show()
```

```python
#Giving label to x-axis and y- axis
x_coordinates=np.array([0,6])
y_coordinates=np.array([0,300])

plt.plot(x_coordinates,y_coordinates)
plt.xlabel("x-coordinate")
plt.ylabel("y-coordinate")
plt.show()
```

```python
#Giving title to graph
x_coordinates=np.array([0,6])
y_coordinates=np.array([0,300])

plt.plot(x_coordinates,y_coordinates)
plt.xlabel("x-coordinate")
plt.ylabel("y-coordinate")
plt.title("Simple line graph")
plt.show()
```

```python
x1=[1,2,3]
y1=[2,4,1]
x2=[1,2,3]
y2=[4,2,1]

plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

#3.BAR GRAPH
Bar graph is a graph by which we can check relationship between a categorical and a continuous variable.
there will be the some gap in between a two bar graph.
for creating a bar graph,we are using a bar function.
by default,bar graph will be the perpendicular of x-axis but we can create a bar graph also which is perpendicular to y-axis.
for creating a bar graph which is parallel to x-axis,we can use barh function

```python
x=np.array(["Ranjan","Apurva","Ganesh","Ankita"])
y=np.array([29,22,23,27])

plt.bar(x,y)
plt.show()
```

#4.HISTOGRAM
Histogram is a type of bar graph which contains no spaces in between a graph.for creating a histogram,we will use hist function.


```python
x=[2,5,5,10,12,70,30,34,45,21,50,67,60,39,40]
range=(0,100)
bins=10

plt.hist(x,bins, range)
plt.show()
```

#5.PIE CHART
Pie chart is the graph of representation of the data in the form of pie.
This is nothing but a circular representation of the data.
For creating a pie chart,we will use pie function.



```python
import matplotlib.pyplot as plt
```

```python
x=[10,8,15,9]
plt.pie(x)
plt.show()
```

```python
x=[10,8,15,9]
mylabels=["a","b","c","d"]
plt.pie(x,labels=mylabels)
plt.show()
```

```python
x=[10,8,15,9]
mylabels=["a","b","c","d"]
plt.pie(x,labels=mylabels,startangle=90)
plt.show()
```

```python
mylabels=["a","b","c","d"]
exp=[0.2,0.1,0.3,0.05]
plt.pie(x,labels=mylabels,startangle=90,explode=exp)
plt.show()
```

```python
mylabels=["a","b","c","d"]
exp=[0.2,0.1,0.3,0.05]
plt.pie(x,labels=mylabels,startangle=90,explode=exp,shadow=True)
plt.show()
```

```python
mylabels=["a","b","c","d"]
exp=[0.2,0.1,0.3,0.05]
plt.pie(x,labels=mylabels,startangle=90,explode=exp,shadow=True,autopct="%0.1f")
plt.show()
```

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
df=seaborn.load_dataset("tips")
```

```python
df.shape
```

```python
df.head()
```

```python
#use swarmplot to visualize tip observation and amounts
#by day of the week
seaborn.swarmplot(x="day",y="tip",data=df)
seaborn.set_style("whitegrid")
plt.show()
```

```python
#visualize tip observations
seaborn.swarmplot(x=df["tip"])
seaborn.set_style("darkgrid")
plt.show()
```

```python
#color points by category
#create customized palette
gender_palette=["#A833FF","#FFAF33"]
seaborn.swarmplot(x="day",y="tip",hue="sex",palette=gender_palette,data=df)
plt.show()
```

```python
#plot tips
seaborn.violinplot(x=df["tip"],color="grey")
plt.show()
```

```python
#draw plot based on variable
seaborn.violinplot(x="sex",y="total_bill",data=df)
plt.show()
```

```python
#split drawings to compare with hue/legend variables
seaborn.violinplot(x="time",y="tip",data=df,hue="sex",palette="dark",split=True)
plt.legend()
plt.show()
```

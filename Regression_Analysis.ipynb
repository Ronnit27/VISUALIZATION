```python
import statsmodels.formula.api as smf
```

```python
model1 = smf.ols('surv ~ imm ',data2).fit().get_robustcov_results(cov_type='HAC',maxlags =18)
```

```python
data2=pd.DataFrame(
    {
        'params':model1.params,
        'pvalues':model1.pvalues
    }
)
```

```python
model1.rsquared
```

```python
model1 = smf.ols('surv ~ imm',data2).fit().get_robustcov_results(cov_type='HAC',maxlags =18)
```
